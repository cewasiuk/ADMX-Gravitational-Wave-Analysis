Measurement Filtered

```python
import h5py
from matplotlib import pyplot as plt
import os 
import numpy as np 

path = os.getcwd()
h5folder = './hr_data'
h5file = 'run1b_nibble5_spectra.h5'
path = h5folder + '/' + h5file
# Open the HDF5 file in read mode
with h5py.File(path, "r") as file:
    # List all groups (datasets and groups) in the file
    print("Keys in the file:", list(file.keys()))
    for key, value in file.attrs.items():
        print(f"{key}: {value}")
    attrs = list(file.attrs.items())
    fstart = float(attrs[0][1])
    fstop = float(attrs[1][1])
    group_name = str(list(file.keys())[0])
    std_dev =  float(attrs[2][1])
    print(std_dev)
    print(group_name)
    if group_name in file:
        data = file[group_name]
        print(f" \n Length of '{group_name}': ", data )



    # yvalues = list(data['measurement_scaled']['yvalues'])
    yvalues = list(data['measurement_filtered']['yvalues'])
    mean = np.average(np.abs(yvalues))
    print(mean)
    yvalues = np.divide(yvalues,mean)
    std_dev = np.divide(std_dev,mean)
    print(len(yvalues))
    f_axis = np.linspace(fstart, fstop, num=len(yvalues), endpoint=True, retstep=False, dtype=None, axis=0)

    x = np.linspace(0, 4000, 100)
plt.figure(figsize=(10, 5))
plt.fill_between(x, -std_dev, std_dev, color='gray', alpha=0.5)
plt.plot(f_axis, yvalues, linestyle = 'none', marker = '.')
plt.hlines(std_dev,0,4000, color = 'black')
plt.title('Filtered Measurement')
plt.hlines(-std_dev,0,4000, color = 'black')
plt.ylabel('Normalized Power')
plt.xlabel('Freq (MHz)')
plt.ylim(-5,5)
plt.xlim(fstart,fstop)
```
